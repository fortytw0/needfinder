```python
import pandas as pd
```

```python
annotation_path = 'data/annotations/needfinding_annotation.xlsx'
keep_columns = ['is_flipped' , 'query' , 'adj_0' , 'Judgment' , 'adj_1']
```

```python
anno_neha_df = pd.read_excel(annotation_path , sheet_name=4 , engine='openpyxl').set_index('original_index')[keep_columns].dropna()
anno_abe_df = pd.read_excel(annotation_path , sheet_name=3 , engine='openpyxl').set_index('original_index')[keep_columns].dropna()
```

```python
anno_neha_df
```

```python
anno_abe_df
```

```python
joined_df = anno_neha_df.join(anno_abe_df, how='outer' , lsuffix='_neha' , rsuffix='_abe')
```

```python
joined_df.head()
joined_df.shape
```

```python
na_df = joined_df[joined_df.isna().any(axis=1)]
na_df.head()
```

```python
anno_df = joined_df.dropna()
print('Dims of anno_df : ' , anno_df.shape )

anno_df.head()
```

```python
anno_df['Judgment_neha'].unique()
```

```python
query_targets = []
```

```python
agreement = anno_df[(anno_df['Judgment_abe']=='<- Left more similar')  & (anno_df['Judgment_neha']=='Right more similar ->') ]
print(agreement.shape)
for query, target in zip(agreement['query_neha'] , agreement['adj_1_neha']) : 

    query_targets.append({'query' : query,
                        'target' : target })
```

```python
agreement = anno_df[(anno_df['Judgment_neha']=='<- Left more similar')  & (anno_df['Judgment_abe']=='Right more similar ->') ]
print(agreement.shape)
for query, target in zip(agreement['query_neha'] , agreement['adj_0_neha']) : 

    query_targets.append({'query' : query,
                        'target' : target })
```

```python
agreement = anno_df[(anno_df['Judgment_neha']=='Both equally similar')  & (anno_df['Judgment_abe']=='Both equally similar') ]
print(agreement.shape)
for query, target1, target2 in zip(agreement['query_neha'] , agreement['adj_0_neha'] , agreement['adj_1_neha']) : 

    query_targets.append({'query' : query,
                        'target' : target1 })

    query_targets.append({'query' : query,
                        'target' : target2 })
```

```python
import json

with open('data/annotations/query_targets.json' , 'w') as f : 
    json.dump(query_targets , f)
```
